### Import libraries

```python
import time
import numpy as numpy
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

## Part 1

### Read txt file and import as pandas Dataframe

```python
inputFile = open('trans.txt','r')
data = [line[:-1].split("\t") for line in inputFile]

te = TransactionEncoder()
te_ary = te.fit(data).transform(data)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

### Run Apriori algorithm

```python
start_time = time.time()
Itemsets = apriori(df, min_support = 0.0001, use_colnames=True, verbose = 1, low_memory = True)
print("Running time: %s seconds" % (round(time.time() - start_time, 2)))
```

### Result output

```python
arr = [0.0005, 0.0004, 0.0003, 0.0002, 0.0001]

def min_freq(feq):
    for i in arr:
        if feq > i:
            return i

final = Itemsets.copy()
final['k'] = [len(i) for i in final['itemsets']]
final['minFreq'] = [min_freq(i) for i in final['support']]
final = final.groupby(['k', 'minFreq']).size().unstack().fillna(0)

for i, j in enumerate(arr):
    if j != 0.0005:
        final[j] += final[arr[i-1]]
```
